```python
import numpy as np
from scipy.linalg import eig
import matplotlib.pyplot as plt

di = 9
hamiltonian_matrix = np.zeros((di, di), dtype=complex)
energys = [[] for i in range(di)]

E_range = np.linspace(0, 5, 100)
alphas = np.arange(di)

for m in range(len(E_range)):
    diagonals = E_range[m] * alphas
    
    for i in range(di):
        hamiltonian_matrix[i, i] = diagonals[i]
        
    hamiltonian_matrix[0, 0] = 0
    
    for i in range(di-1):
        hamiltonian_matrix[i, i+1] = hamiltonian_matrix[i+1, i] = 10

    w, v = eig(hamiltonian_matrix)
    w_sorted_indices = np.argsort(w)
    
    for i in range(di):
        energys[i].append(w[w_sorted_indices[i]])

# Plotting the results
for i in range(di):
    plt.plot(E_range, [np.real(e) for e in energys[i]], label=f"State {i+1}")

plt.xlabel('Electric field')
plt.ylabel('Energy')
plt.legend()
plt.show()
```
